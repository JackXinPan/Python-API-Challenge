```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
import warnings

# Import API key
from api_keys import g_key

# Access maps with unique API key
gmaps.configure(api_key=g_key)
```

### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
#load import weather data and make it into DataFrame
csv = "Output/weather_data.csv"
weather_data = pd.read_csv(csv)
weather_data.head()
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# Access maps with unique API key
gmaps.configure(api_key=g_key)
# Store latitude and longitude in locations
locations = weather_data[["Lat", "Lng"]]
# defining weight for heatmap
humidity = weather_data["Humidity"]
```

```python
# Plot Heatmap
humidity_heatmap = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=humidity, 
                                 dissipating=False, max_intensity=100,
                                 point_radius=2)


# Add layer
humidity_heatmap.add_layer(heat_layer)

# Display figure
humidity_heatmap
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
# the ideal weather conditions are:
#  A max temperature lower than 26 degrees celsius but higher than 21.
#  Wind speed less than 10 mph.
#  Zero cloudiness.
#  Humidity less than 85% 
#subset to include data points with only ideal conditions
weather_ideal = weather_data.loc[(weather_data["Max Temp"] <= 26) &
                                (weather_data["Max Temp"] >= 21) & 
                                (weather_data["Wind Speed"] <= 10) &
                                 (weather_data["Humidity"] <= 85) &
                                 (weather_data["Cloudiness"] == 0),:] 
#display ideal cities
weather_ideal
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
#storing variable in to hotel_df
hotel_df = weather_ideal

#inserting new blank column
hotel_df["Hotel Name"] = ""

#clean output
warnings.filterwarnings('ignore')

#output
hotel_df
```

```python
# find hotels near ideal locations
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
#setting up param dict for for loop
params = {
    "radius": 5000,
    "keyword": "hotel",
    "key": g_key,
}
print("Starting Search")
print('------------------------------------------------')
# use iterrows to iterate through pandas dataframe
for index, row in hotel_df.iterrows():
    #assembling appropriate coordinates for params
    lat = row["Lat"]
    lng = row["Lng"]
    city = row["City"]
    params["location"] = f'{lat},{lng}'
    
    # assemble url and make API request
    response = requests.get(base_url, params=params).json()
    
    # extract results
    results = response['results']
    # error catch if there are no hotels in range or missing
    try:
        print(f"The first hotel result near {city} is: {results[0]['name']}.")
        print("------------------------------------------------")
        hotel_df.loc[index, "Hotel Name"] = results[0]['name']
    #what to output if there is an error    
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
        print("------------------------------------------------")
print("Finished Search")
#clean output
warnings.filterwarnings('ignore')
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Lat", "Lng"]]
```

```python
# Add marker layer ontop of heat map
markers = gmaps.marker_layer(locations, info_box_content = hotel_info)
humidity_heatmap.add_layer(markers)

# Display figure
humidity_heatmap
```
