```python
import requests
import numpy as np
import time
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import json
from api_keys import weather_api_key
# look for the closest city using geo coordinates
from citipy import citipy
```

## Generate Cities List

```python
#creating lists for for loop
cities = []
loops = ['']*1600

#loop to find cities 
for loop in loops:
    #create random set of geo coordinates
    lat = np.random.randint(-90,90)
    lng = np.random.randint(-180,180)
    #citipy to find nearest city to random set of geo coordinates
    city = citipy.nearest_city(lat, lng).city_name
    #prevent duplicate cities on list when appending
    if city not in cities:
        cities.append(city)

#number of cities to validate city list
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Build partial query URL
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "metric"
key = weather_api_key
query_url = f"{url}appid={key}&units={units}&q="
```

```python
#lists for data weather data retrieval
city_name = []
lat = []
lng = []
max_temp = []
humidity = []
cloudiness = []
wind_speed = []
country = []
date = []
# setting indexing counters for for loop
record = 0
set_page = 1

#using print and for loop to create output of lists of weather stats for each valid city
print("Beginning Data Retrieval ")
print("-------------------------------")
for city in cities:
    #in case of Index or Type error if city has missing data
    try:
        response = requests.get(query_url + city).json()
        city_name.append(response['name'])
        lat.append(response['coord']['lat'])
        lng.append(response['coord']['lon'])
        max_temp.append(response['main']['temp_max'])
        humidity.append(response['main']['humidity'])
        cloudiness.append(response['clouds']['all'])
        wind_speed.append(response['wind']['speed'])
        country.append(response['sys']['country'])
        date.append(response['dt'])
        #for indexing responses
        if record > 49:
            record = 1
            set_page = set_page + 1
        else:
            record = record + 1
            
        print(f'Processing record {record} of Set {set_page} | {city}')
    #exception that occurs if city has missing data
    except(KeyError, IndexError):
              print("City not found. Skipping...")
print("-------------------------------")
print("Data Retrieval Complete")
print("-------------------------------")
```

```python
#checking number of valid cities to make sure it it exceeds 500
len(city_name)
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#create dataframe of weather statistics of recorded cities
weather_data = pd.DataFrame({
    "City": city_name, "Lat": lat, "Lng": lng,
    "Max Temp": max_temp, "Humidity": humidity,
    "Cloudiness": cloudiness, "Wind Speed": wind_speed,
    "Country": country, "Date": date
})

#save to output
weather_data.to_csv("Output/weather_data.csv", index=False)

#output
weather_data.head()
```

```python
#descriptive statistics of weather data
weather_descriptive_stats = weather_data.describe()
weather_descriptive_stats.to_csv("Output/weather_descriptive_stats.csv", index=False)

#output
weather_descriptive_stats
```

## Linear Regression

```python
#subsetting weather data into northern and southern hemisphere
north_data = weather_data.loc[weather_data["Lat"] >= 0, :]
south_data = weather_data.loc[weather_data["Lat"] <= 0, :]

#defining x variables for all scatterplots for regression
north_lat = north_data["Lat"]
south_lat = south_data["Lat"]
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
#defining y variable
north_temp = north_data["Max Temp"]

# Calculate and print the correlation coefficient 
correlation = st.pearsonr(north_lat, north_temp)

print(f"The R value is : {correlation[0]}.")
print(f"The R-square value is : {correlation[0]*correlation[0]}.")

#linear regressions model
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(north_lat, north_temp)
regress_values = north_lat * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.annotate(line_eq,(0,-10),fontsize=15,color="red")
plt.plot(north_lat,regress_values,"r-")

#scatter plot and labeling
plt.scatter(north_lat, north_temp, marker="o", facecolors="b", edgecolors="black", alpha=0.75)
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (C)")
plt.title("Latitudes in Northern Hemisphere Vs Max Temperature(C) 27/03/2021")

#save to output
plt.savefig("Output/North_Latitude_Max_Temp_Regression.png")

#output
plt.show()
```

## 2nd Observable Trend Latitude vs Humidity Regression

From both the output of the Northern and Southern Hemisphere Latitudes vs Humidity Regression codes, we can observe that as latitude increases so does humidity. In more detail, past the equator, as latitude increases by 1 degree, humidity will rise by 0.15%. The R value or correlation coefficient posits a weak positive correlation. While in the southern hemisphere, as latitude increases by 1 degree humidity will increase by 0.25% with a weak to moderate positive correlation.

One would hypothesise that humidity would be higher in warmer climates/temperatures. With warmer climates usually existing near the equator as previously noted. Therefore, the relationship observed from the southern hemisphere regression is expected, but not with the northern hemisphere. As latitude increases while inside the northern hemisphere, you are leaving the equator and should expect humidity to fall because of it. An unobserved weather shock on the day of data collection could explain the unexpected observation, or more likely a symptom of the biased sample selection method when finding cities. The algorithm to find cities does bias towards coastal cities because randomised geo coordinates are most likely to hit sea. Coastal cities do logically have more humidity compared to landlocked cities.
